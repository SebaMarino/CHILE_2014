```python
# K-Fold Cross Validation function
def KFoldCrossValidate(model,target,n_folds=10,names=None,output=False):
    skf = StratifiedKFold(target.ravel(), n_folds=n_folds)
    predicted = []
    actual = []
    iteration=1
    for train_index, test_index in skf:
        feat_train, feat_test = features[train_index], features[test_index]
        target_train, target_test = target[train_index], target[test_index]
        
        model.fit(feat_train,target_train)
        print "Iteration: {}".format(iteration)
        if output:
            print "Feature Importances:"
            for i in xrange(len(columns)):
                print "  {}: {}".format(columns[i],model.feature_importances_[i])
            print "OOB Score: {}".format(model.oob_score_)
            print "Test Score: {}".format(model.score(feat_test,target_test))
        actual.append(target_test.ravel())
        predicted.append(model.predict(feat_test).ravel())
        iteration += 1
        results = pd.crosstab(np.concatenate(actual), np.concatenate(predicted), rownames=["Actual"], colnames=["Predicted"])
        if names is not None:
            results.rename(index=names, columns=names, inplace=True)
    return results

# This class is a utility used to pass only numeric values through the DataFrameMapper
class NullScaler(object):
    @staticmethod
    def fit(X):
        return X
    @staticmethod
    def transform(X):
        return X
```

```python
# Calculate the average color from multiple measurement columns, ignoring nans
def calc_avg_color(src):
    val = []
    for col in src.values:
        if (np.isnan(col).sum() > 1):
            val.append(np.nan)
        else:
            val.append(col[~np.isnan(col)].mean())
    return val

# Calculate B-R from multiple measurement columns, ignoring measurements where one or both are nans
def calc_BR(src):
    BR = []
    for col in src.values:
        blue = col[0::2]
        red = col[1::2]
        if (np.isnan(blue).sum() > 1) |(np.isnan(red).sum() > 1):
            BR.append(np.nan)
        else:
            BR_ = blue-red
            BR.append(BR_[~np.isnan(BR_)].mean())
    return BR
```

```python
# QSO selection tool. Used to derive the QSO/non-QSO features.
# http://butler.lab.asu.edu/qso_selection/
import qso_fit_fix

# function for calculating the Stetson J variability index
def stetson_j(MJD, mag, err):
    nobs = len(MJD);
    magi = mag[0:nobs-1]
    erri = err[0:nobs-1]
    magj = mag[1:nobs]
    errj = err[1:nobs]

    weights = np.ones(magi.shape)   

    meanmag = np.tile(np.mean(mag), nobs-1 );
    deltai = np.sqrt(nobs/(nobs-1))*(magi-meanmag)/erri;
    deltaj = np.sqrt(nobs/(nobs-1))*(magj-meanmag)/errj;
    Pk = deltai*deltaj;

    Jind = np.sum(weights*np.sign(Pk)*np.sqrt(np.abs(Pk)))/np.sum(weights);
    return Jind

# Function for calculating the Median Absolute Dispersion
def mad(data, axis=None):
    return np.median(np.absolute(data - np.median(data, axis)), axis)

# Function returns statistics for lightcurves used as features for the classifiers.
def get_stats(data, label, values, band_filter='r'):
    data_stats = defaultdict(float)
    data_stats['ID'] = []
    data_stats['mag_kurtosis'] = []
    data_stats['mag_skew'] = []
    data_stats['mag_mad'] = []
    data_stats['chi2_nu'] = []
    data_stats['chi2qso_nu'] = []
    data_stats['j_ind'] = []
    for i, val in enumerate(values):
        data_stats['ID'].append(val)
        data_stats['mag_kurtosis'].append(sp.kurtosis(data[data[label] == val].Mag.values))
        data_stats['mag_skew'].append(sp.skew(data[data[label] == val].Mag.values))
        data_stats['mag_mad'].append(mad(data[data[label] == val].Mag.values))
        val_time = data[data[label] == val].Time.values
        val_mag = data[data[label] == val].Mag.values
        val_err = data[data[label] == val].Err.values
        qso = qso_fit_fix.qso_fit(val_time, val_mag, val_err, filter=band_filter)
        data_stats['chi2_nu'].append(qso['chi2/nu'])
        data_stats['chi2qso_nu'].append(qso['chi2_qso/nu'])
        data_stats['j_ind'].append(stetson_j(val_time, val_mag, val_err))
        if np.mod(i, 100) == 0:
            print "{} percent processed".format(float(i) / float(len(values)) * 100)
    print data_stats.keys()
    return pd.DataFrame(data_stats,columns=data_stats.keys()).set_index('ID')
```

```python
# Generalized random forest function used for all classifiers in the project
def group2_random_forests(data, col_features, col_target, filter_limit=0):
    # Get target types and apply filter limit
    data = data.copy()
    types = data[col_target].values.tolist()
    print 'N:', len(data)
    counts = defaultdict(int)
    for i in types:
        counts[i] += 1
    if filter_limit > 0:
        main_types = sorted([key for key, value in counts.items() if value > filter_limit])
    else:
        main_types = sorted(set(types))
    print 'Classes ({}): {}'.format(len(counts), counts)
    print 'Filtered Classes ({}): {}'.format(len(main_types), main_types)
    data = data[data[col_target].isin(main_types)]
    
    
    # Split data into features and target
    data_features = data[col_features]
    data_target = data[[col_target]]
    
    # map the features and target into a form we can use for the random forest process
    feature_mapper = [(x, NullScaler()) for x in col_features]
    mapper = DataFrameMapper(feature_mapper)
    features = mapper.fit_transform(data_features)
    
    mapper = DataFrameMapper([(col_target, sklearn.preprocessing.LabelEncoder())])
    target = mapper.fit_transform(data_target)
    
    # Split into test and train sets
    feat_train, feat_test, target_train, target_test = train_test_split(features, target, test_size=0.3)
        
    # Build the classifier
    rf = RandomForestClassifier(n_estimators=100,oob_score=True,compute_importances=True,n_jobs=1, )
    rf.fit(feat_train,target_train)
    print "Feature Importances:"
    for i in xrange(len(col_features)):
        print "  {}: {}".format(col_features[i],rf.feature_importances_[i])
    print "\nOOB Score: {}".format(rf.oob_score_)
    #print "Test Data Score:", rf.score(feat_test,target_test)
    
    # Calculate and display results on test set
    results = pd.crosstab(target_test.ravel(), rf.predict(feat_test).ravel(), rownames=["Actual"], colnames=["Predicted"])
    names = {}
    for i, name in enumerate(sorted(main_types)):
        names[i] = name
    results.rename(index=names, columns=names, inplace=True)
    results.columns.name = 'Predicted'
    #print
    #print "Percent Correct: {}".format(float(results.values.diagonal().sum()) / float(results.values.sum())) 
    #print results
    
    return (rf, results, features, target, names)
```

```python
# Function returns prediction and probabilities of test set after applying random forest classifier
def apply_rf_model(model, test, names):
    prob = model.predict_proba(test)
    pred = model.predict(test)
    results = pd.DataFrame(test.index.copy())
    results.rename(columns={0:'ID'},inplace=True)
    for i, name in enumerate(sorted(names.values())):
        results[name] = prob[:,i]
    results['Class_Pred'] = [names[x] for x in pred]
    return results
```

```python
# Import LightCurves data
import glob
fnames = glob.glob("LightCurves/*")
dfs = []
search = '(?<=_)(.+)(?=_[0-9])'
for i, fname in enumerate(fnames):
    with open(fname, 'r') as f:
        lines = f.readlines()
    data = [line.strip().split() for line in lines]
    data = pd.DataFrame(data[2::], columns=data[0][1::], dtype=float)
    data['field'] = re.search(search, fname).group()
    dfs.append(data)
    
    if np.mod(i,1000) == 0:
        print "{} of {}".format(i, len(fnames))
lightcurves = pd.concat(dfs, axis=0, join='outer', ignore_index=False)

# Add a column with unique source ID from CCD, FOV, and source number
lightcurves['ID'] = lightcurves.apply(lambda x:'%.4d_%s' % (int(x['source']),x['field']),axis=1)

# Add column with adjusted lightcurve flux using transparency data
lightcurves['flux_adj'] = lightcurves.flux / lightcurves.transparency

# Add column with adjusted lightcurve flux using transparency data
lightcurves['rad_adj'] = lightcurves.radius / lightcurves.transparency


# Add calculated column for flux error
gain = 4.47
lightcurves['e_flux_adj'] = np.sqrt(lightcurves['flux_adj']/gain + lightcurves['sigma']**2)
lightcurves['mag'] = -2.5*np.log10(lightcurves['flux_adj'])
lightcurves['e_mag'] = 2.5/np.log(10)*lightcurves['e_flux_adj']/lightcurves['flux_adj']

lightcurves.head()
```

```python
# Filter out bad flux and radius data
lightcurves = lightcurves[lightcurves.flux != 0]
lightcurves = lightcurves[lightcurves.radius > 0]
lightcurves
```

```python
def get_class_pred(df, src, tgt):
    return df.apply(lambda x: df.drop(src,axis=1).columns[x.argmax()] if x[src] == tgt else x[src],axis=1)
```

```python
# Propagate Periodic probabilities
temp_df = results_EvP.drop('EvP_Class_Pred', axis=1)
temp_df = temp_df.apply(lambda x: np.asarray(x) * np.asarray(results_NP.Periodic))
results_combined = results_NP.drop(['Periodic'], axis=1).join(temp_df)
results_combined['Class_Pred'] = results_combined.apply(\
    lambda x: results_combined.drop('NP_Class_Pred',axis=1).columns[x.argmax()+1] \
    if x['NP_Class_Pred'] == 'Periodic' else x['NP_Class_Pred'],axis=1)
results_combined = results_combined.drop(['NP_Class_Pred'], axis=1)
results_combined.head()
```

```python
# Propagate Eclipsing probabilities
temp_df = results_E.drop('E_Class_Pred', axis=1)
temp_df = temp_df.apply(lambda x: np.asarray(x) * np.asarray(results_combined.E))
results_combined = results_combined.drop(['E'], axis=1).join(temp_df)
results_combined['Class_Pred'] = results_combined.apply(\
    lambda x: results_combined.drop('Class_Pred',axis=1).columns[x.argmax()+1] \
    if x['NP_Class_Pred'] == 'E' else x['NP_Class_Pred'],axis=1)
results_combined = results_combined.drop(['NP_Class_Pred'], axis=1)
results_combined.head()
```

```python
# Propagate Pulsating probabilities
temp_df = results_P.drop('P_Class_Pred', axis=1)
temp_df = temp_df.apply(lambda x: np.asarray(x) * np.asarray(results_combined.P))
results_combined = results_combined.drop(['P'], axis=1).join(temp_df)
results_combined.head()
```

```python
# Calculate final class prediction and if it meets threshold
threshold = 0.80
results_full = results_combined.copy()
results_full['Class_Pred'] = results_full.apply(lambda x: results_full.columns[x.argmax()], axis = 1)
results_full['Meets_Threshold'] = results_full.drop('Class_Pred',axis=1).apply(lambda x: max(x) > threshold, axis = 1)
HTML(results_full.head().to_html())
```

```python
breakdown = pd.DataFrame({'All': results_full.Class_Pred.value_counts(), 
                          'Over Threshold': results_full[results_full.Meets_Threshold].Class_Pred.value_counts(),
                          'Under Threshold': results_full[~results_full.Meets_Threshold].Class_Pred.value_counts()})
print "Classifications meeting confidence threshold ({}%)".format(threshold * 100)
print results_full.Meets_Threshold.value_counts()
# print "\nClass breakdown:"
# print results_full.Class_Pred.value_counts()
# print "\nOver threshold class breakdown:"
# print results_full[results_full.Meets_Threshold].Class_Pred.value_counts()
# print "\nUnder threshold class breakdown:"
# print results_final[~results_full.Meets_Threshold].Class_Pred.value_counts()
breakdown.replace(to_replace=np.nan, value=0, inplace=True)
breakdown.sort('All', ascending=False)
```

```python
# Output data for sources below confidence threshold
columns = ['field','source','RA','DEC','P1','P2','P3','dP1','dP2','dP3','Conf1','Conf2','Conf3']
results_out = results_full[~results_full.Meets_Threshold].drop(['Class_Pred','Meets_Threshold'], axis=1).join(sources[columns])
results_out.to_csv('SourceList.csv')
```
